3. What is the difference between discrete and continuous distributions?
- A discrete distribution describes the probability of outcomes of a discrete random variable, which takes countable values (e.g., 0, 1, 2). Common examples include the Binomial and Poisson distributions.

A continuous distribution represents probabilities for a continuous random variable, which takes uncountably infinite values within an interval (e.g., 2.5, 3.14). Examples include the Normal and Exponential distributions.

Discrete distributions use a probability mass function (PMF), while continuous ones use a probability density function (PDF).
4. What are probability distribution functions (PDF)?
- A Probability Distribution Function (PDF) describes how the probability is distributed over the values of a continuous random variable.

5. How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?
- The PDF (Probability Density Function) shows how the probability is distributed over values of a continuous random variable, giving the likelihood near a point. It does not give exact probabilities at a point.

The CDF (Cumulative Distribution Function) gives the total probability that a random variable is less than or equal to a value, and always ranges from 0 to 1. PDF is the derivative of the CDF, while the CDF is the integral of the PDF.

13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?
- The Central Limit Theorem (CLT) states that the sampling distribution of the sample mean approaches a normal distribution as the sample size increases, regardless of the population‚Äôs distribution, provided the samples are independent and identically distributed.

It is critical because it allows statisticians to use normal distribution techniques (like confidence intervals and hypothesis tests) even when the original data is not normally distributed.

14. How does the Central Limit Theorem relate to the normal distribution?
- The Central Limit Theorem explains that, as sample size increases, the distribution of sample means becomes approximately normal, even if the population distribution is not.

This links directly to the normal distribution, allowing statisticians to apply normal-based methods (like Z-tests) to analyze sample data from any population.

```python
#3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution?
def bernoulli_pdf(x, p):
    """
    Calculate the PDF of a Bernoulli distribution.

    Parameters:
    x (int): The outcome (0 or 1)
    p (float): The probability of success (0 <= p <= 1)

    Returns:
    float: Probability of the outcome x
    """
    if x not in [0, 1]:
        raise ValueError("x must be 0 or 1 for a Bernoulli distribution.")
    if not (0 <= p <= 1):
        raise ValueError("p must be between 0 and 1.")

    return p if x == 1 else 1 - p
```

```python
#9.Write a Python function to calculate Z-scores from a dataset and plot them?
import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_z_scores(data):
    """
    Calculate Z-scores from a dataset and plot them.

    Parameters:
    data (list or array-like): Input numerical dataset
    """
    # Convert to NumPy array
    data = np.array(data)

    # Calculate mean and standard deviation
    mean = np.mean(data)
    std = np.std(data)

    # Compute Z-scores
    z_scores = (data - mean) / std

    # Plot Z-scores
    plt.figure(figsize=(8, 5))
    plt.plot(z_scores, marker='o', linestyle='', color='darkblue')
    plt.axhline(0, color='red', linestyle='--', label='Mean (Z=0)')
    plt.axhline(1, color='green', linestyle='--', label='1 Std Dev')
    plt.axhline(-1, color='green', linestyle='--')
    plt.title('Z-scores of Dataset')
    plt.xlabel('Index')
    plt.ylabel('Z-score')
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.legend()
    plt.show()

    return z_scores
```

```python
#10.Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution?
import numpy as np
import matplotlib.pyplot as plt

def demonstrate_clt(distribution, dist_params, sample_sizes, num_samples=1000):
    """
    Demonstrates the Central Limit Theorem using the specified distribution.

    Parameters:
    distribution (callable): NumPy random distribution function (e.g., np.random.exponential)
    dist_params (dict): Parameters for the distribution
    sample_sizes (list): List of sample sizes to test (e.g., [5, 30, 100])
    num_samples (int): Number of samples to draw for each size
    """
    plt.figure(figsize=(15, 4))

    for i, n in enumerate(sample_sizes, 1):
        sample_means = []
        for _ in range(num_samples):
            sample = distribution(size=n, **dist_params)
            sample_means.append(np.mean(sample))

        # Plot histogram of sample means
        plt.subplot(1, len(sample_sizes), i)
        plt.hist(sample_means, bins=30, density=True, alpha=0.7, color='skyblue', edgecolor='black')
        plt.title(f'Sample Size = {n}')
        plt.xlabel('Sample Mean')
        plt.ylabel('Density')
        plt.grid(True, linestyle='--', alpha=0.6)

    plt.suptitle('Central Limit Theorem Demonstration using Exponential Distribution', fontsize=14)
    plt.tight_layout(rect=[0, 0, 1, 0.95])
    plt.show()

# Example: Exponential distribution (non-normal)
demonstrate_clt(distribution=np.random.exponential, dist_params={'scale': 2.0}, sample_sizes=[5, 30, 100])
```

```python
#15.Simulate multiple samples from a normal distribution and verify the Central Limit Theorem?
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def clt_normal_demo(mu=50, sigma=10, sample_sizes=[5, 30, 100], num_samples=1000):
    """
    Simulates multiple samples from a normal distribution and verifies the CLT
    by plotting sampling distributions of the sample mean.

    Parameters:
    - mu (float): Mean of the original normal distribution
    - sigma (float): Standard deviation of the original distribution
    - sample_sizes (list): Different sample sizes to demonstrate CLT
    - num_samples (int): Number of samples to draw for each sample size
    """
    plt.figure(figsize=(15, 4))

    for i, n in enumerate(sample_sizes, 1):
        sample_means = []

        for _ in range(num_samples):
            sample = np.random.normal(loc=mu, scale=sigma, size=n)
            sample_means.append(np.mean(sample))

        # Plot histogram of sample means
        plt.subplot(1, len(sample_sizes), i)
        plt.hist(sample_means, bins=30, density=True, alpha=0.6, color='lightgreen', edgecolor='black')

        # Overlay theoretical normal distribution
        x = np.linspace(min(sample_means), max(sample_means), 100)
        theoretical_pdf = norm.pdf(x, loc=mu, scale=sigma / np.sqrt(n))
        plt.plot(x, theoretical_pdf, 'r-', label='Theoretical PDF')

        plt.title(f'Sample Size = {n}')
        plt.xlabel('Sample Mean')
        plt.ylabel('Density')
        plt.legend()
        plt.grid(True, linestyle='--', alpha=0.6)

    plt.suptitle('CLT Verification with Normal Distribution', fontsize=14)
    plt.tight_layout(rect=[0, 0, 1, 0.95])
    plt.show()

# Run the demo
clt_normal_demo()

```

```python
#17. Generate random variables and calculate their corresponding probabilities using the binomial distribution?
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

def binomial_simulation(n=10, p=0.5, size=1000):
    """
    Generate random variables from a binomial distribution and calculate their probabilities.

    Parameters:
    - n (int): Number of trials
    - p (float): Probability of success
    - size (int): Number of samples to generate
    """
    # Generate random variables
    samples = np.random.binomial(n=n, p=p, size=size)

    # Count occurrences of each outcome
    values, counts = np.unique(samples, return_counts=True)

    # Calculate empirical probabilities
    empirical_probs = counts / size

    # Calculate theoretical probabilities using binomial PMF
    x = np.arange(0, n+1)
    theoretical_probs = binom.pmf(x, n, p)

    # Plotting
    plt.figure(figsize=(8, 5))
    plt.bar(values, empirical_probs, width=0.6, color='skyblue', edgecolor='black', label='Empirical PMF')
    plt.plot(x, theoretical_probs, 'ro-', label='Theoretical PMF')

    plt.title(f'Binomial Distribution (n={n}, p={p})')
    plt.xlabel('Number of Successes')
    plt.ylabel('Probability')
    plt.xticks(range(n+1))
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.legend()
    plt.show()

    # Print example probabilities
    print("k\tEmpirical P(X=k)\tTheoretical P(X=k)")
    for k, emp, theo in zip(x, np.bincount(samples, minlength=n+1)/size, theoretical_probs):
        print(f"{k}\t{emp:.4f}\t\t\t{theo:.4f}")

# Run the simulation
binomial_simulation(n=10, p=0.5, size=10000)
```

```python
#18.5 Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution?
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_z_score(x, mu, sigma):
    """
    Calculate the Z-score for a given data point.
    """
    return (x - mu) / sigma

def plot_standard_normal_with_z(z_score):
    """
    Plot the standard normal distribution and shade the area up to the given Z-score.
    """
    x = np.linspace(-4, 4, 1000)
    y = norm.pdf(x, 0, 1)

    plt.figure(figsize=(8, 5))
    plt.plot(x, y, label='Standard Normal PDF', color='blue')
    plt.fill_between(x, 0, y, where=(x <= z_score), color='skyblue', alpha=0.5, label=f'P(Z ‚â§ {z_score:.2f})')
    plt.axvline(z_score, color='red', linestyle='--', label=f'Z = {z_score:.2f}')

    plt.title('Standard Normal Distribution with Z-score')
    plt.xlabel('Z')
    plt.ylabel('Density')
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.legend()
    plt.show()

# Example inputs
x_value = 75
mean = 70
std_dev = 5

# Calculate and plot
z = calculate_z_score(x_value, mean, std_dev)
print(f"Z-score for x = {x_value}: {z:.2f}")
plot_standard_normal_with_z(z)

```

```python
#19.5 Implement hypothesis testing using Z-statistics for a sample dataset?
import numpy as np
from scipy.stats import norm

def z_test(sample, mu_0, sigma, alpha=0.05):
    """
    Perform a one-sample Z-test.

    Parameters:
    - sample (array-like): Sample data
    - mu_0 (float): Hypothesized population mean (H0)
    - sigma (float): Population standard deviation
    - alpha (float): Significance level (default 0.05)

    Returns:
    - Z-score, p-value, decision
    """
    n = len(sample)
    sample_mean = np.mean(sample)

    # Compute Z-score
    z = (sample_mean - mu_0) / (sigma / np.sqrt(n))

    # Two-tailed test p-value
    p_value = 2 * (1 - norm.cdf(abs(z)))

    # Decision
    decision = "Reject H0" if p_value < alpha else "Fail to Reject H0"

    # Output
    print(f"Sample Mean = {sample_mean:.2f}")
    print(f"Z-score = {z:.3f}")
    print(f"p-value = {p_value:.4f}")
    print(f"Decision: {decision} (at Œ± = {alpha})")

    return z, p_value, decision

# Example: Create a sample from a normal population
np.random.seed(42)
sample_data = np.random.normal(loc=52, scale=5, size=30)  # Sample from N(52, 5^2)

# Test if mean is significantly different from 50, assuming known œÉ = 5
z_test(sample=sample_data, mu_0=50, sigma=5)
```

```python
#20.Create a confidence interval for a dataset using Python and interpret the result?
import numpy as np
import scipy.stats as stats

def confidence_interval(data, confidence=0.95):
    """
    Calculate the confidence interval for the mean of a dataset.

    Parameters:
    - data (array-like): The sample data
    - confidence (float): Confidence level (default 95%)

    Returns:
    - (lower_bound, upper_bound)
    """
    n = len(data)
    mean = np.mean(data)
    std_err = stats.sem(data)  # Standard error
    t_critical = stats.t.ppf((1 + confidence) / 2.0, df=n-1)
    margin_of_error = t_critical * std_err

    lower = mean - margin_of_error
    upper = mean + margin_of_error

    print(f"Sample Mean = {mean:.2f}")
    print(f"{int(confidence * 100)}% Confidence Interval: ({lower:.2f}, {upper:.2f})")

    return lower, upper

# Example: Generate sample data
np.random.seed(1)
sample_data = np.random.normal(loc=100, scale=15, size=30)

# Compute 95% confidence interval
confidence_interval(sample_data)

```

```python
#21 Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean?
import numpy as np
import scipy.stats as stats

# Step 1: Generate data from a normal distribution
np.random.seed(42)
data = np.random.normal(loc=70, scale=10, size=50)  # Mean = 70, SD = 10, n = 50

# Step 2: Calculate the 95% confidence interval for the mean
def calculate_confidence_interval(data, confidence=0.95):
    n = len(data)
    mean = np.mean(data)
    std_err = stats.sem(data)  # Standard error of the mean
    t_critical = stats.t.ppf((1 + confidence) / 2, df=n - 1)
    margin_of_error = t_critical * std_err
    lower = mean - margin_of_error
    upper = mean + margin_of_error

    print(f"Sample Mean = {mean:.2f}")
    print(f"{int(confidence*100)}% Confidence Interval: ({lower:.2f}, {upper:.2f})")
    return lower, upper

# Run the function
calculate_confidence_interval(data)
```

```python
#22.Write a Python script to calculate and visualize the probability density function (PDF) of a normal distribution?
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters for the normal distribution
mu = 0      # Mean
sigma = 1   # Standard deviation

# Generate x values over a range
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)

# Calculate the PDF using scipy's norm.pdf
pdf = norm.pdf(x, loc=mu, scale=sigma)

# Plot the PDF
plt.figure(figsize=(8, 5))
plt.plot(x, pdf, label='Normal PDF', color='blue')
plt.fill_between(x, pdf, color='skyblue', alpha=0.5)
plt.axvline(mu, color='red', linestyle='--', label='Mean (Œº)')
plt.title('Probability Density Function of Normal Distribution')
plt.xlabel('x')
plt.ylabel('Density')
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.show()
```

```python
#23.Use Python to calculate and interpret the cumulative distribution function (CDF) of a Poisson distribution?
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Set lambda (mean rate of occurrences)
lam = 3  # You can change this to experiment

# Define a range of x values
x = np.arange(0, 15)

# Calculate the CDF
cdf_values = poisson.cdf(x, mu=lam)

# Print CDF values
for xi, cdf in zip(x, cdf_values):
    print(f"P(X ‚â§ {xi}) = {cdf:.4f}")

# Plot the CDF
plt.figure(figsize=(8, 5))
plt.step(x, cdf_values, where='post', label=f'Poisson CDF (Œª = {lam})', color='blue')
plt.scatter(x, cdf_values, color='red')
plt.title('CDF of Poisson Distribution')
plt.xlabel('x')
plt.ylabel('P(X ‚â§ x)')
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.show()
```

```python
#24. Simulate a random variable using a continuous uniform distribution and calculate its expected value?
import numpy as np

# Define the interval [a, b]
a = 10
b = 20
n = 10000  # Number of samples

# Simulate random variables from Uniform(a, b)
samples = np.random.uniform(low=a, high=b, size=n)

# Calculate theoretical expected value
expected_theoretical = (a + b) / 2

# Calculate empirical (sample) mean
expected_empirical = np.mean(samples)

# Print results
print(f"Theoretical Expected Value E[X] = {(a + b) / 2:.2f}")
print(f"Empirical Mean from Simulated Data = {expected_empirical:.2f}")
```

```python
#25. Write a Python program to compare the standard deviations of two datasets and visualize the difference?
import numpy as np
import matplotlib.pyplot as plt

# Generate two datasets (you can also use real data)
np.random.seed(0)
data1 = np.random.normal(loc=50, scale=5, size=100)   # Mean=50, SD=5
data2 = np.random.normal(loc=50, scale=10, size=100)  # Mean=50, SD=10

# Calculate standard deviations
std1 = np.std(data1, ddof=1)  # Sample standard deviation
std2 = np.std(data2, ddof=1)

# Print standard deviations
print(f"Standard Deviation of Dataset 1: {std1:.2f}")
print(f"Standard Deviation of Dataset 2: {std2:.2f}")

# --------- Visualization ---------

# Bar chart of standard deviations
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.bar(['Dataset 1', 'Dataset 2'], [std1, std2], color=['skyblue', 'salmon'])
plt.title('Standard Deviation Comparison')
plt.ylabel('Standard Deviation')
plt.grid(axis='y', linestyle='--', alpha=0.6)

# Histograms of both datasets
plt.subplot(1, 2, 2)
plt.hist(data1, bins=20, alpha=0.7, label='Dataset 1 (œÉ‚âà5)', color='skyblue', edgecolor='black')
plt.hist(data2, bins=20, alpha=0.7, label='Dataset 2 (œÉ‚âà10)', color='salmon', edgecolor='black')
plt.title('Histogram of Datasets')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```

```python
#26.Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution?
import numpy as np

# Step 1: Generate data from a normal distribution
np.random.seed(42)
data = np.random.normal(loc=100, scale=15, size=1000)  # Mean=100, SD=15, n=1000

# Step 2: Calculate Range
range_value = np.max(data) - np.min(data)

# Step 3: Calculate IQR
q1 = np.percentile(data, 25)
q3 = np.percentile(data, 75)
iqr = q3 - q1

# Print results
print(f"Range: {range_value:.2f}")
print(f"Q1 (25th percentile): {q1:.2f}")
print(f"Q3 (75th percentile): {q3:.2f}")
print(f"IQR (Q3 - Q1): {iqr:.2f}")
```

```python
#27.Implement Z-score normalization on a dataset and visualize its transformation?
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate sample data
np.random.seed(0)
data = np.random.normal(loc=50, scale=10, size=1000)  # Original data

# Step 2: Z-score normalization
mean = np.mean(data)
std = np.std(data)
z_scores = (data - mean) / std

# Step 3: Plot original vs normalized data
plt.figure(figsize=(12, 5))

# Original data histogram
plt.subplot(1, 2, 1)
plt.hist(data, bins=30, color='skyblue', edgecolor='black')
plt.title('Original Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.axvline(mean, color='red', linestyle='--', label=f'Mean = {mean:.2f}')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)

# Normalized data histogram
plt.subplot(1, 2, 2)
plt.hist(z_scores, bins=30, color='salmon', edgecolor='black')
plt.title('Z-score Normalized Data')
plt.xlabel('Z-score')
plt.ylabel('Frequency')
plt.axvline(0, color='red', linestyle='--', label='Mean = 0')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()

# Optional: print basic stats
print(f"Original Mean: {mean:.2f}, Std Dev: {std:.2f}")
print(f"Normalized Mean: {np.mean(z_scores):.2f}, Std Dev: {np.std(z_scores):.2f}")
```

```python
#28. Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution?

import numpy as np
from scipy.stats import skew, kurtosis

def calculate_skewness_kurtosis(size=1000, mean=0, std_dev=1):
    """
    Generate a dataset from a normal distribution and compute skewness and kurtosis.

    Parameters:
    - size (int): Number of data points
    - mean (float): Mean of the normal distribution
    - std_dev (float): Standard deviation of the normal distribution

    Returns:
    - skewness (float), kurtosis (float)
    """
    # Generate random data from a normal distribution
    data = np.random.normal(loc=mean, scale=std_dev, size=size)

    # Calculate skewness and excess kurtosis (normal dist: skew=0, kurt=0)
    skewness = skew(data)
    excess_kurtosis = kurtosis(data)  # By default, scipy returns excess kurtosis

    # Print results
    print(f"Generated {size} samples from N({mean}, {std_dev}¬≤)")
    print(f"Skewness: {skewness:.4f}")
    print(f"Excess Kurtosis: {excess_kurtosis:.4f} (Normal = 0)")

    return skewness, excess_kurtosis

# Example usage
calculate_skewness_kurtosis()
```
